```python
import scipy.io

anno_train = scipy.io.loadmat("./datasets/annotations/anno_train.mat")
anno_val = scipy.io.loadmat("./datasets/annotations/anno_val.mat")
```

```python
anno_train.keys()
anno_val.keys()
```

```python
anno_train["anno_train_aligned"]
anno_val["anno_val_aligned"]
```

```python
anno_train["anno_train_aligned"][0][0][0][0]
```

```python
datasets = []

for cell in anno_train["anno_train_aligned"][0]:
    city_name, im_name, bbs = cell[0][0]
    bbs_new = []
    for each in bbs:
        class_label, x1, y1, w, h, instance_id, x1_vis, y1_vis, w_vis, h_vis = each
        bbs_new.append(
            {
                "class_label": class_label,
                "x1": x1,
                "y1": y1,
                "w": w,
                "h": h,
                "instance_id": instance_id,
                "x1_vis": x1_vis,
                "y1_vis": y1_vis,
                "w_vis": w_vis,
                "h_vis": h_vis,
            }
        )
    datasets.append({"city_name": city_name[0], "im_name": im_name[0], "bbs": bbs_new})
    # print(city_name, im_name, bbs)
# len(datasets[2]["bbs"])
```
